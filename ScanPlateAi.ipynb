```python
from ultralytics import YOLO
model = YOLO(r"C:\Users\Alim Haciverdiyev\ScanPlateAi\runs\detect\train\weights\best.pt")
img_path = r"C:\Users\Alim Haciverdiyev\ScanPlateAi\personal-test\IMG-20250603-WA0021.jpg"
results = model(img_path)
results[0].show()
```

```python
import cv2
import easyocr
from matplotlib import pyplot as plt
import re

# Orijinal resmi oku
image = cv2.imread(img_path)

# Kutuları ve sınıf etiketlerini al
boxes = results[0].boxes.xyxy.cpu().numpy().astype(int)           # Kutu koordinatları
classes = results[0].boxes.cls.cpu().numpy().astype(int)          # Sınıf index'leri
names = model.names                                               # Sınıf adları (örneğin {0: 'plate', 1: 'audi', ...})

# OCR hazırlığı
reader = easyocr.Reader(['en', 'tr'])

# Sadece "plate" sınıfı olanları işle
for box, cls in zip(boxes, classes):
    class_name = names[cls]
    if class_name != "plate":
        continue  # Sadece "plate" sınıfı için devam et

    x1, y1, x2, y2 = box

    # Plaka bölgesini kırp
    cropped = image[y1:y2, x1:x2]

    # Çözünürlüğü artır
    upscale_factor = 4
    resized = cv2.resize(cropped, (0, 0), fx=upscale_factor, fy=upscale_factor, interpolation=cv2.INTER_CUBIC)

    # Görüntüyü göster (isteğe bağlı)
    plt.imshow(cv2.cvtColor(resized, cv2.COLOR_BGR2RGB))
    plt.title("Yüksek Çözünürlüklü Plaka")
    plt.axis("off")
    plt.show()

    # OCR ile plaka metnini oku
    result = reader.readtext(resized)

    # Tüm sonucu birleştir
    full_text = " ".join([text for (_, text, _) in result])

    # İlk rakamdan itibaren plaka verisini çek
    match = re.search(r"\d.*", full_text)
    plate_text = match.group(0) if match else ""

    print(f"📸 Okunan Plaka: {plate_text}")
```
